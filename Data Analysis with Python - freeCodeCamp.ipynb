```python
def calculate(list):
    calculations = {}
    if len(list) == 9: 
        my_list = np.reshape(list,(3,3))
        df = pd.DataFrame(my_list, columns={'C1','C2','C3'})
        mean = [df.mean().tolist(),df.T.mean().tolist(), my_list.flatten().mean()]
        var = [df.var().tolist(),df.T.var().tolist(), my_list.flatten().var()]
        std = [df.std().tolist(),df.T.std().tolist(), my_list.flatten().std()]
        maximum = [df.max().tolist(),df.T.max().tolist(), my_list.flatten().max()]
        minimum = [df.min().tolist(),df.T.min().tolist(), my_list.flatten().min()]
        s = [df.sum().tolist(),df.T.sum().tolist(), my_list.flatten().sum()]
        calculations['mean'] = mean
        calculations['variance'] = var
        calculations['standard deviation'] = std
        calculations['maximum'] = maximum
        calculations['minimum'] = minimum
        calculations['sum'] = s

        return calculations
    else: 
        return "List must contain nine numbers."
a = [0,1,2,3,4,5,6,7,8]
calculate(a)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Import data (Make sure to parse dates. Consider setting index column to 'date'.)
df = pd.read_csv('fcc-forum-pageviews.csv')
# Clean data
df = df.sort_values(ascending=True, by='value')
df = df.iloc[33:1271]
df['day'] = df['date'].apply(lambda x: pd.to_datetime(x).strftime("%d"))
df['date'] = df['date'].apply(lambda x: pd.to_datetime(x).strftime("%Y-%m"))
    # Save image and return fig (don't change this part)
df = df.groupby(['date','day'], as_index=False).sum()
```

```python
df
```

```python
import datetime
```

```python
fig = df_line.plot(figsize=(32,8), color='red')
plt.title('Daily freeCodeCamp Forum Page Views 5/2019-12/2019', fontsize=16, color='black')
plt.xlabel('Date', fontsize=14, color='black')
plt.ylabel('Page Views', fontsize=14, color='black')
plt.show()
# plt.xlim(datetime.date(2016,7,1), datetime.date(2020,1,1))
```

```python
# show average daily page views for each month grouped by year
df['month'] = df['date'].apply(lambda x: pd.to_datetime(x).strftime("%m"))
df['year'] = df['date'].apply(lambda x: pd.to_datetime(x).strftime("%Y"))
```

```python
df = df.groupby(['year','month'],as_index=False).mean()
```

```python
jan = df[df['month'] == '01']['value'].tolist()
feb = df[df['month'] == '02']['value'].tolist()
mar = df[df['month'] == '03']['value'].tolist()
apr = df[df['month'] == '04']['value'].tolist()
may = df[df['month'] == '05']['value'].tolist()
jun = df[df['month'] == '06']['value'].tolist()
jul = df[df['month'] == '07']['value'].tolist()
aug = df[df['month'] == '08']['value'].tolist()
sep = df[df['month'] == '09']['value'].tolist()
octo = df[df['month'] == '10']['value'].tolist()
nov = df[df['month'] == '11']['value'].tolist()
dec = df[df['month'] == '12']['value'].tolist()
jan.insert(0,0)
feb.insert(0,0)
mar.insert(0,0)
apr.insert(0,0)
year = [2016,2017,2018,2019]
```

```python
df_bar = pd.DataFrame({'January': jan,
                   'February': feb, 'March':mar,'April': apr,'May':may,'June':jun,
                      'July': jul,'August':aug,'September':sep, 'October':octo,'November':nov,
                      'December':dec}, index=year)
```

```python
df_bar
```

```python
fig = df_bar.plot.bar(figsize=(12,8))
plt.xlabel('Years',fontsize=12, color="black")
plt.ylabel('Averager Page Views',fontsize=12, color="black")
plt.legend(title='Month')
fig = plt.show()
```
